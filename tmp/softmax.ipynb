```python
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
# Gaussian distribution using NumPy
mu = 0.0
sigma = 1.0
x = np.linspace(-4, 4, 1000)
coeff = 1.0 / (sigma * np.sqrt(2 * np.pi))
pdf_np = coeff * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

plt.figure(figsize=(7, 4))
plt.plot(x, pdf_np, label=f"NumPy N({mu}, {sigma}^2)")
plt.xlabel("x")
plt.ylabel("density")
plt.title("Gaussian PDF (NumPy)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
# Gaussian distribution using PyTorch (apples-to-apples)
mu_t = torch.tensor(0.0)
sigma_t = torch.tensor(1.0)
x_t = torch.linspace(-4.0, 4.0, 1000)
coeff_t = 1.0 / (sigma_t * torch.sqrt(torch.tensor(2.0 * np.pi)))
pdf_t = coeff_t * torch.exp(-0.5 * ((x_t - mu_t) / sigma_t) ** 2)

plt.figure(figsize=(7, 4))
plt.plot(x_t.detach().cpu().numpy(), pdf_t.detach().cpu().numpy(), label=f"PyTorch N({mu_t.item()}, {sigma_t.item()}^2)")
plt.xlabel("x")
plt.ylabel("density")
plt.title("Gaussian PDF (PyTorch)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
